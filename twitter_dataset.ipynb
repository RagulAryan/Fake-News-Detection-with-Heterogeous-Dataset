```python
import re
def clean_tweets(text):
    text=re.sub("RT @[\w]*:","",text)
    text=re.sub("@[\w]*:","",text)
    text=re.sub("http?://[A-Za-z0-9./]*","",text)
    text=re.sub("\n","",text)
    return text
```

```python
tweet_df['text']=tweet_df['text'].apply(lambda x: clean_tweets(x))
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')
```

```python
tweet_df['text'].apply(lambda x: [print("\tText : {}, Entity : {}".format(ent.text,ent.label_)) if(not ent.text.startswith('#')) else "" for ent in nlp(x).ents])  
```

```python
tweet_df['entities']=tweet_df['text'].apply(lambda x:[(ent.text,ent.label_) if (not ent.text.startswith('#')) else "" for ent in nlp(x).ents])
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
sid=SentimentIntensityAnalyzer()
```

```python
tweet_df['sentiment']=tweet_df['text'].apply(lambda x:sid.polarity_scores(x))
```

```python
tweet_df.head(10)
```

```python
analysis_df=pd.concat([tweet_df[['text','user','location','entities']],tweet_df['sentiment'].apply(pd.Series)],axis=1)
```

```python
import matplotlib.pyplot as plt
def plot_learning_curve(negative,positive):
    
    plt.plot(neg)
    plt.plot(pos)
    
    plt.ylabel("Range")
    plt.xlabel("Sentiments")
    plt.legend(["neg","pos"],loc ="upper left")
    plt.show()
plot_learning_curve(neg,pos)
```
